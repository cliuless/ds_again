```python
#http://web.mta.info/developers/turnstile.html
#get data data from 12/30/2017 to present
def getData(weeks):
    urls ="http://web.mta.info/developers/data/nyct/turnstile/turnstile_{}.txt"
    merged_dfs = []    
    for week in weeks:
        upload_url = urls.format(week)
        csv_file = pd.read_csv(upload_url)
        merged_dfs.append(csv_file)
    return pd.concat(merged_dfs, ignore_index = True) #resets index in order to avoid multiples of same index   
        
```

#### Challenge 3
These counts are for every n hours. (What is n?) We want total daily entries.
Now make it that we again have the same keys, but now we have a single value for a single day, which is the total number of passengers that entered through this turnstile on this day.

```python
#computing the difference in column entries 
entry_df['Diff_ENTRIES']=entry_df.groupby([entry_df.T_H.dt.date, entry_df.STATION, entry_df.SCP])['ENTRIES'].diff() 
entry_df  
```

```python
#dropping negative values
df_intervalcount= entry_df[(entry_df.Diff_ENTRIES >=0)]
df_intervalcount
```

```python
#check what to filter on upper boundary
df_intervalcount['Diff_ENTRIES'].describe().astype(int)
```

```python
#set upper boundary as cutoff at 1000000
df_intervalcount= df_intervalcount[(df_intervalcount.Diff_ENTRIES <=1000000)]
df_intervalcount
```

```python
#Answer 1.3
#daily count of every turnstile by station
daily_count = df_intervalcount.groupby([df_intervalcount.DATE, df_intervalcount.STATION, df_intervalcount.SCP], as_index = False)['Diff_ENTRIES'].sum()
daily_count.rename(index=str, columns={"Diff_ENTRIES": "DAILY_TOTAL"}, inplace=True)
daily_count
```

#### Challenge 5
So far we've been operating on a single turnstile level, let's combine turnstiles in the same ControlArea/Unit/Station combo. There are some ControlArea/Unit/Station groups that have a single turnstile, but most have multiple turnstilea-- same value for the C/A, UNIT and STATION columns, different values for the SCP column.
We want to combine the numbers together -- for each ControlArea/UNIT/STATION combo, for each day, add the counts from each turnstile belonging to that combo.

```python
# UNIT is irrelevant since its the same for every exit within a station
daily_count_station_date = daily_count.groupby([daily_count.DATE, daily_count.STATION], as_index = False)['DAILY_TOTAL'].sum()
```

```python
#ANSWER CHALLENGE 5
#Grouped the data according to date, station and daily_total
daily_count_station_date.DAILY_TOTAL = daily_count_station_date.DAILY_TOTAL.astype(int)
daily_count_station_date
```
