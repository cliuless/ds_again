### **Challenge Number 1**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data"
df = pd.read_csv(url, header = None, names = ['Party'] + ['v'+str(i) for i in range(1,17)])
def numify(i):
    if i =='y':
        return 1
    elif i =='n':
        return 0
    else:
        return np.nan
def numify1(i):
    if i == 'republican':
        return 0
    elif i == 'democrat':
        return 1
    else:
        return np.nan
for i in range(1,17):
    df['v'+str(i)] = [numify(j) for j in df['v'+str(i)]]
df.fillna(df.mean(), inplace = True)
from sklearn.cross_validation import train_test_split
y = df['Party']
y = [numify1(x) for x in y]
X = df[['v' + str(i) for i in range(1,17)]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .30, random_state = 4444)
import sklearn.metrics as metrics
from sklearn.neighbors import KNeighborsClassifier
acck = []
preck = []
reck = []
effonek = []
for i in range(20):
    knn = KNeighborsClassifier(n_neighbors = i+1)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acck.append(metrics.accuracy_score(y_test, y_pred))
    preck.append(metrics.precision_score(list(y_test), list(y_pred)))
    reck.append(metrics.recall_score(y_test, y_pred))
    effonek.append(metrics.f1_score(y_test, y_pred))
acck = np.array(acck)
preck = np.array(preck)
reck = np.array(reck)
effonek = np.array(effonek)
print('K nearest neighbors, k from 1 to 20:')
print('best ks for accuracy: ', end = '')
print([x + 1 for x in np.where(acck==acck.max())])
print('best accuracy: ', end = '')
print(acck.max())
print('All Accuracies: ', end = '')
print(acck)
print('best ks for precision: ', end = '')
print([x + 1 for x in np.where(preck==preck.max())])
print('best precision: ', end = '')
print(preck.max())
print('All Precisions: ', end = '')
print(preck)
print('best ks for recall: ', end = '')
print([x + 1 for x in np.where(reck==reck.max())])
print('best recall: ', end = '')
print(reck.max())
print('All Recalls: ', end = '')
print(reck)
print('best ks for f1: ', end = '')
print([x + 1 for x in np.where(effonek==effonek.max())])
print('best f1: ', end = '')
print(effonek.max())
print('All f1s: ', end = '')
print(effonek)
print('Logistic Regression: ')
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=4444)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print('Accuracy: ', end = '')
print(metrics.accuracy_score(y_test, y_pred))
print('Precision: ', end = '')
print(metrics.precision_score(y_test, y_pred))
print('Recall: ', end = '')
print(metrics.recall_score(y_test, y_pred))
print('F1: ', end = '')
print(metrics.f1_score(y_test, y_pred))
print('Gaussian Naive Bayes: ')
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
print('Accuracy: ', end = '')
print(metrics.accuracy_score(y_test, y_pred))
print('Precision: ', end = '')
print(metrics.precision_score(y_test, y_pred))
print('Recall: ', end = '')
print(metrics.recall_score(y_test, y_pred))
print('F1: ', end = '')
print(metrics.f1_score(y_test, y_pred))
print('Support Vector Machine: ')
from sklearn.svm import SVC
sv = SVC(random_state = 4444)
sv.fit(X_train, y_train)
y_pred = sv.predict(X_test)
print('Accuracy: ', end = '')
print(metrics.accuracy_score(y_test, y_pred))
print('Precision: ', end = '')
print(metrics.precision_score(y_test, y_pred))
print('Recall: ', end = '')
print(metrics.recall_score(y_test, y_pred))
print('F1: ', end = '')
print(metrics.f1_score(y_test, y_pred))
print('Decision Tree: ')
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier(random_state = 4444)
DTC.fit(X_train, y_train)
y_pred = DTC.predict(X_test)
print('Accuracy: ', end = '')
print(metrics.accuracy_score(y_test, y_pred))
print('Precision: ', end = '')
print(metrics.precision_score(y_test, y_pred))
print('Recall: ', end = '')
print(metrics.recall_score(y_test, y_pred))
print('F1: ', end = '')
print(metrics.f1_score(y_test, y_pred))
print('Random Forest: ')
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(random_state = 4444)
RF.fit(X_train, y_train)
y_pred = RF.predict(X_test)
print('Accuracy: ', end = '')
print(metrics.accuracy_score(y_test, y_pred))
print('Precision: ', end = '')
print(metrics.precision_score(y_test, y_pred))
print('Recall: ', end = '')
print(metrics.recall_score(y_test, y_pred))
print('F1: ', end = '')
print(metrics.f1_score(y_test, y_pred))
```

```python
def ratify(x, i):
    if(x==i):
        return 1
    else:
        return 0
df2 = pd.read_csv('2013_movies.csv')
df2.dropna(inplace=True)
del df2['ReleaseDate']
del df2['Director']
del df2['Title']
df2['Budget'] = (df2.Budget - min(df2.Budget))/(max(df2.Budget)-min(df2.Budget))
df2['DomesticTotalGross'] = (df2.DomesticTotalGross - min(df2.DomesticTotalGross))/(max(df2.DomesticTotalGross)-min(df2.DomesticTotalGross))
df2['Runtime'] = (df2.Runtime - min(df2.Runtime))/(max(df2.Runtime)-min(df2.Runtime))
y2 = df2['Rating']
y2PG = [ratify(x, 'PG') for x in y2]
y2PG13 = [ratify(x, 'PG-13') for x in y2]
y2R = [ratify(x, 'R') for x in y2]
classes = [y2PG, y2PG13, y2R]
cname = ['PG: ', 'PG-13: ', 'R: ']
X2 = df2[[v for v in df2.columns.values if v != 'Rating']]
tscore = ['precision', 'recall']
for i in classes:
    print(cname[classes.index(i)])
    for score in tscore:
        print(score+': ')
        print('KNN (k=4): ', end = '')
        print(np.mean(cross_val_score(KNeighborsClassifier(n_neighbors = 4), X2, i, scoring = 'precision', cv = 5)))
        print('Logistic Regression: ', end = '')
        print(np.mean(cross_val_score(LogisticRegression(random_state=4444), X2, i, scoring = 'recall', cv = 5)))
        print(' ')
    print(' ')
```

```python
def binify(x):
    if(x==1):
        return 0
    else:
        return 1
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/haberman/haberman.data'
df3 = pd.read_csv(url, header = None, names = ['Age','Year','Nodes','Status'])
W = df3[['Age', 'Year', 'Nodes']]
z = df3['Status']
z = [binify(x) for x in z]
W_train, W_test, z_train, z_test = train_test_split(W, z, test_size = .30, random_state = 4444)
WminAge = min(W_train['Age'])
WmaxAge = max(W_train['Age'])
WminYear = min(W_train['Year'])
WmaxYear = max(W_train['Year'])
WminNodes = min(W_train['Nodes'])
WmaxNodes = max(W_train['Nodes'])
W_train['Age'] = [(x-WminAge)/(WmaxAge-WminAge) for x in W_train['Age']]
W_train['Year'] = [(x-WminYear)/(WmaxYear-WminYear) for x in W_train['Year']]
W_train['Nodes'] = [(x-WminNodes)/(WmaxNodes-WminNodes) for x in W_train['Nodes']]
W_test['Age'] = [(x-WminAge)/(WmaxAge-WminAge) for x in W_test['Age']]
W_test['Year'] = [(x-WminYear)/(WmaxYear-WminYear) for x in W_test['Year']]
W_test['Nodes'] = [(x-WminNodes)/(WmaxNodes-WminNodes) for x in W_test['Nodes']]
lr = LogisticRegression(random_state=4444)
lr.fit(W_train, z_train)
z_score=lr.predict_proba(W_test)[:,1]
fpr, tpr,_ = roc_curve(z_test, z_score)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot([0,1],[0,1], label = 'random guessing')
plt.plot(fpr,tpr, label = 'roc curve')
plt.legend(loc = 'lower right')
plt.xlabel('FPR')
plt.title('ROC curve for Logistic Regression')
plt.ylabel('TPR');
print('Logistic Regression AUC: '+str(roc_auc))
```
