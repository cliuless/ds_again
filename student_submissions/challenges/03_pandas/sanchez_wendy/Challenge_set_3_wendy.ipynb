#### **Challenge 1**

Plot domestic total gross over time.


```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df = pd.read_csv('../../challenges_data/2013_movies.csv',
                 parse_dates=['ReleaseDate'], infer_datetime_format=True)
```

```python
df.shape
```

```python
dom_tot_gross_time= df.sort_values(['ReleaseDate','DomesticTotalGross'])[['ReleaseDate','Title', 'DomesticTotalGross']]
dom_tot_gross_time.reset_index()
```

```python
plt.figure(figsize=(20,10))
ax=dom_tot_gross_time.plot(x= "ReleaseDate",y ="DomesticTotalGross" , title= "Domestic Total Gross Over Time", grid=True)
ax.set_xlabel('ReleaseDate')
ax.set_ylabel("Domestic Total Gross")
```

#### **Challenge 6**

Bin your dataset into months and make a bar graph of the mean domestic total gross by month.  Error bars will represent the standard error of the mean.

Title of graph should include:  Mean Domestic Total Gross by Month in 2013

Topic for consideration:  what is the correct formula for the standard error of the mean?  Examine the error bars and see if they are "reasonable."


```python
df.ReleaseDate = df.ReleaseDate.dt.month
```

```python
release_domestic = df.groupby(['ReleaseDate'])['DomesticTotalGross'].mean()
sd=df.groupby(['ReleaseDate'])['DomesticTotalGross'].sem()
```

```python

plt.figure(figsize=(20,10))
ax=release_domestic.plot.bar(x= "ReleaseDate",y ="DomesticTotalGross" , title= "Mean Domestic Total Gross by Month in 2013", yerr =sd)
ax.set_xlabel('Month')
ax.set_ylabel("Domestic Total Gross")
```
