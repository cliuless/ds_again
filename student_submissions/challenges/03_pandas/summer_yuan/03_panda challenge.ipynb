Challenge1: Plot domestic total gross over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
moviecsv =  pd.read_csv('2013_movies.csv')
df = pd.DataFrame(moviecsv)
df.head()
```

```python
df_sorted = df.sort_values(by = 'ReleaseDate')
```

```python
totalgross  = df_sorted['DomesticTotalGross']
releaseDate = pd.to_datetime(df_sorted['ReleaseDate'])
```

```python
plt.plot_date(releaseDate, totalgross)
plt.title('Domestic Total Gross over release date')
plt.xlabel('Release Date')
plt.ylabel('Domestic Total Gross')
plt.xticks(rotation = 45)
```

```python
dfsorted = df 
dfsorted['ReleaseDate'] = pd.to_datetime(df['ReleaseDate'])
dfsorted = df.sort_values(by =['ReleaseDate'])
```

```python
#divide df into four data frames based on rating, (try to do this in pd pivot table)
dfg= dfsorted.loc[df['Rating'] =='G']
dfpg= dfsorted.loc[df['Rating'] == 'PG']
dfpg13 = dfsorted.loc[df['Rating'] == 'PG-13']
dfr = dfsorted.loc[df['Rating'] == 'R']
```

```python
plt.figure(figsize= (12,3))
G= plt.plot_date(dfg['ReleaseDate'],dfg['DomesticTotalGross'],label ='G')
PG = plt.plot_date(dfpg['ReleaseDate'],dfpg['DomesticTotalGross'],label ='PG')
PG13 = plt.plot_date (dfpg13['ReleaseDate'], dfpg13['DomesticTotalGross'],label = 'PG13')
R= plt.plot_date(dfr['ReleaseDate'],dfr['DomesticTotalGross'], label = 'R')
plt.xticks (rotation = 45)
plt.xlabel('Release date')
plt.ylabel('Domestic total gross')
plt.title('Domestic total gross on Release date')
plt.legend(loc = 'upper left')
#how to show legend in only one dot ??
```

```python
dfsorted.head()
```

```python
#get the month in each release date
dfsorted['month'] = pd.DatetimeIndex(dfsorted['ReleaseDate']).month
dfsorted.head()
```

```python
#group by month 
dfbymonth = dfsorted.groupby('month')['DomesticTotalGross']
mean = dfbymonth.mean()
stdev = dfbymonth.std()
n = dfbymonth.count()
errormean = stdev/np.sqrt(n)
```

```python
mean.plot(yerr = errormean, kind = 'bar', figsize = (10,6), style = 'y')        
plt.title('Mean Domestic Total Gross by Month in 2013')
plt.ylabel('Domestic Total Gross (in $100m)')
```
