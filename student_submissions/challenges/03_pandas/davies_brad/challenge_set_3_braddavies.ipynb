```python
#Topic:        Challenge Set 3
#Subject:      Explored movie data
#Date:         01/22/2018
#Name:         Brad Davies
#Worked with:  None


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('2013_movies.csv')
```

```python
df.sort_values(by='ReleaseDate')
```

```python
#Challenge 1:
df_1 = df.groupby('ReleaseDate').mean().reset_index()
df_1.plot(x='ReleaseDate', y='DomesticTotalGross', title='Challenge 1')
```

```python
#Challenge 2:
df_2 = df.groupby('Runtime').mean().reset_index()
df_2.plot(x='Runtime', y='DomesticTotalGross', title = 'Challenge 2')
```

```python
#Challenge 3:

df_3 = df.groupby('Rating').mean()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax2 = ax.twinx()

df_3['Runtime'].plot(kind='bar', color='red', ax=ax, position=0, width=0.25)
df_3['DomesticTotalGross'].plot(kind='bar', color='blue', ax=ax2, position=1, width=0.25)
ax.set_ylabel = ('Runtime')
ax2.set_ylabel = ('Gross')

plt.show()
```

```python
#Challenge 4:

df_4_g_temp = df[df['Rating']=='G']
df_4_pg_temp = df[df['Rating']=='PG']
df_4_pg13_temp = df[df['Rating']=='PG-13']
df_4_r_temp = df[df['Rating']=='R']

df_4_g = df_4_g_temp.groupby('ReleaseDate').mean().reset_index()
df_4_pg = df_4_pg_temp.groupby('ReleaseDate').mean().reset_index()
df_4_pg13 = df_4_pg13_temp.groupby('ReleaseDate').mean().reset_index()
df_4_r = df_4_r_temp.groupby('ReleaseDate').mean().reset_index()
```

```python
df_4_g.plot(x='ReleaseDate', y='DomesticTotalGross', title='G')
df_4_pg.plot(x='ReleaseDate', y='DomesticTotalGross', title='PG')
df_4_pg13.plot(x='ReleaseDate', y='DomesticTotalGross', title='PG-13')
df_4_r.plot(x='ReleaseDate', y='DomesticTotalGross', title='R')
```

```python
#Challenge 6

def getMonth(x):
    
    if x[1] == '/':
        return x[0]
    else:
        return x[0:2]

df_6_temp = df
df_6_temp['Month'] = list(map(getMonth, df_6_temp['ReleaseDate']))
```

```python
df_6 = df_6_temp.groupby('Month').mean().reset_index()
df_6.plot(x='Month', y='DomesticTotalGross', title='Challenge 6')
```
