```python
import requests
from bs4 import BeautifulSoup

gmail_url="https://mail.google.com"
soup=BeautifulSoup(requests.get(gmail_url).text,"lxml")
print(soup.prettify())
```

### When do we want to use selenium?

To handle dynamic web pages. ie: when the content we want to crawl is being added to the page via JavaScript, rather than baked into the HTML.

### XPATH: 

We can also use XPATH to select html elements from a page

[nice ref here](http://selenium-interview-questions.blogspot.in/2014/02/how-to-identify-elements-using-xpath-in.html)

### Scraping boxofficemojo with selenium

```python
matrix_url = "http://www.boxofficemojo.com/movies/?id=matrix.htm"
driver.get(matrix_url)
```

```python
# 'contains' will find a match on the text, in this case return b tag
# 'contains' allows us to use a substring to find our match
# this example: we are grabbing text from the upper right corner..

gross_selector = '//font[contains(text(), "Domestic")]/b'
print(driver.find_element_by_xpath(gross_selector).text)
```

```python
# scraping genre (we find 4 elements that match this xpath, we'll return b
genre_selector = '//a[contains(@href, "/genres/chart/")]/b'
for genre_anchor in driver.find_elements_by_xpath(genre_selector):
    print(genre_anchor.text)
```

```python
# selecting specific value of 2000
inf_adjust_2000_selector = '//select[@name="ticketyr"]/option[@value="2017"]'
driver.find_element_by_xpath(inf_adjust_2000_selector).click()
```

```python
go_button = driver.find_element_by_name("Go")
go_button.click()
```

Now the page has changed, it's showing inflation adjusted numbers. We can grab the new, adjusted number

```python
gross_selector = '//font[contains(text(), "Domestic ")]/b'
print(driver.find_element_by_xpath(gross_selector).text)
```
